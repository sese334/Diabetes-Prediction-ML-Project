### Diabetes Prediction Dataset Overview

The **Diabetes Prediction Dataset** encapsulates a wealth of medical and demographic insights gleaned from patient records, coupled with their diabetes status categorized as either positive or negative. This rich reservoir of information encompasses crucial attributes including age, gender, body mass index (BMI), hypertension, heart disease, smoking history, HbA1c level, and blood glucose level.

#### Potential Applications

- **Healthcare Advancement**: This dataset serves as a formidable foundation for constructing sophisticated machine learning models aimed at prognosticating diabetes onset based on comprehensive medical histories and demographic profiles. By leveraging these models, healthcare professionals can effectively discern patients at heightened risk of diabetes, facilitating early intervention and personalized treatment strategies.

- **Research Endeavors**: Beyond its clinical utility, researchers can harness the dataset to delve into the intricate interplay between diverse medical and demographic variables and the propensity for diabetes development. Such explorations promise to unearth invaluable insights into the multifaceted nature of diabetes etiology, thereby fostering groundbreaking advancements in preventive medicine and public health initiatives.

> Here check count of **Gender**, **Hypertension**, **Heart Disease**, **smoking_history** and **Diabetes** columns

```python
# Function to add counts on bars
def add_counts(ax):
    for p in ax.patches:
        ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                    textcoords='offset points')

# Set up the matplotlib figure
fig, axes = plt.subplots(3, 2, figsize=(15, 15))

# Plot gender grouped by diabetes
ax = sns.countplot(ax=axes[0, 0], x='gender', hue='diabetes', data=df)
axes[0, 0].set_title('Gender grouped by Diabetes')
add_counts(ax)

# Plot hypertension grouped by diabetes
ax = sns.countplot(ax=axes[0, 1], x='hypertension', hue='diabetes', data=df)
axes[0, 1].set_title('Hypertension grouped by Diabetes')
add_counts(ax)

# Plot heart disease grouped by diabetes
ax = sns.countplot(ax=axes[1, 0], x='heart_disease', hue='diabetes', data=df)
axes[1, 0].set_title('Heart Disease grouped by Diabetes')
add_counts(ax)

# Plot smoking history grouped by diabetes
ax = sns.countplot(ax=axes[1, 1], x='smoking_history', hue='diabetes', data=df)
axes[1, 1].set_title('Smoking History grouped by Diabetes')
add_counts(ax)

# Plot diabetes
ax = sns.countplot(ax=axes[2, 0], x='diabetes', data=df)
axes[2, 0].set_title('Diabetes Count')
add_counts(ax)

# Create pie plot for diabetes
diabetes_counts = df['diabetes'].value_counts()
axes[2, 1].pie(diabetes_counts, labels=diabetes_counts.index, autopct='%1.1f%%', startangle=90)
axes[2, 1].set_title('Diabetes Distribution')
axes[2, 1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
axes[2, 1].legend(title='Diabetes:', loc='upper right')
# Adjust the layout
plt.tight_layout()

# Show the plots
plt.show()
```

# <b> <span style='color:#ff7f0e'>Step 10</span> <span style='color:red'>|</span> <span style='color:#2ca02c ;font-style: italic;animation: pulse 2s infinite;'>Principal Component Analysis (PCA) </span></b>

```python


# Assume df is your dataframe

# Selecting features and target variable
features = ['gender', 'age', 'hypertension', 'heart_disease', 'smoking_history', 'bmi', 'HbA1c_level', 'blood_glucose_level']
X = df[features]
y = df['diabetes']

# Standardizing the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Applying PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Plotting the cumulative explained variance
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1),
         pca.explained_variance_ratio_.cumsum(), marker='o', linestyle='--')
plt.title('Explained Variance by Number of Principal Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid()

# Find the index of the maximum cumulative explained variance
max_index = pca.explained_variance_ratio_.cumsum().argmax()

# Annotate the point with the highest cumulative explained variance
plt.annotate(f'Max: PC {max_index + 1}',
             xy=(max_index + 1, pca.explained_variance_ratio_.cumsum()[max_index]),
             xytext=(max_index + 2, pca.explained_variance_ratio_.cumsum()[max_index] - 0.05),
             arrowprops=dict(facecolor='black', arrowstyle='->', color='black'))

plt.show()

# Printing explained variance ratios
for i, ratio in enumerate(pca.explained_variance_ratio_.cumsum()):
    print(f'Principal Component {i+1}: {ratio:.4f} cumulative explained variance')

# Choose the number of components that explain most of the variance
n_components = max_index + 1

# Applying PCA with the optimal number of components
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X_scaled)

```

> Best Number of Component is 8

```python
# Splitting the data into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)



# Initializing and training the XGBoost model

xgb_model = XGBClassifier(random_state=42)

xgb_model.fit(X_train, y_train)



# Making predictions

y_pred = xgb_model.predict(X_test)



# Evaluating the model

accuracy = accuracy_score(y_test, y_pred)

print(f'XGBoost Accuracy: {accuracy:.4f}')

print(f'XGBoost Classification Report:\n{classification_report(y_test, y_pred)}')
```

# Handling Class Imbalance using XGBoost Scale Weights"

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 1. Splitting the dataset
# stratify=y ensures that the positive/negative diabetes ratio is the same in train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X_pca, y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

# 2. Initializing XGBoost with class weight balancing

xgb_model = XGBClassifier(
    random_state=42,
    scale_pos_weight=2,       # <-- modified parameter to balance positive class
    eval_metric='logloss'     # suppresses evaluation warnings
)

# Training the model
xgb_model.fit(X_train, y_train)

# 3. Making predictions
y_pred = xgb_model.predict(X_test)

# 4. Evaluating model performance
accuracy = accuracy_score(y_test, y_pred)
print(f'XGBoost Accuracy: {accuracy:.4f}')
print(f'Classification Report:\n{classification_report(y_test, y_pred)}')
```

```python
from sklearn.ensemble import StackingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report

# 1. Base Models (Level-0 Learners)
# Each model is adjusted to handle class imbalance
estimators = [
    ('xgb', XGBClassifier(
        scale_pos_weight=2,        # balances minority class inside XGBoost
        eval_metric='logloss',
        random_state=42
    )),
    ('rf', RandomForestClassifier(
        class_weight='balanced',   # automatic weight balancing for Random Forest
        random_state=42
    ))
]

# 2. Meta Model (Level-1 Learner)
# Logistic Regression combines decisions from the base models
clf = StackingClassifier(
    estimators=estimators,
    final_estimator=LogisticRegression()
)

print("Training Stacking Ensemble Model...")
clf.fit(X_train, y_train)

# 3. Predictions & Evaluation
y_pred_stack = clf.predict(X_test)

print(f"Stacking Ensemble Accuracy: {accuracy_score(y_test, y_pred_stack):.4f}")
print(f"Classification Report:\n{classification_report(y_test, y_pred_stack)}")
```

```python
from lightgbm import LGBMClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# 1. Train–Test Split (same logic used previously)
# If you are working on original data, replace X_pca with X
X_train, X_test, y_train, y_test = train_test_split(
    X_pca, y,
    test_size=0.2,
    random_state=42,
    stratify=y               # keeps class ratio consistent in train/test sets
)

# 2. LightGBM Model (Fast + High Performance)
lgbm_model = LGBMClassifier(
    n_estimators=200,        # number of boosting trees (200 is a strong starting point)
    learning_rate=0.05,      # lower LR for stable learning and higher accuracy
    scale_pos_weight=2,      # handles class imbalance (try 2, 3, or 4)
    random_state=42,
    verbosity=-1             # suppress warnings/log spam
)

print("Training LightGBM Model...")
lgbm_model.fit(X_train, y_train)

# 3. Predictions
y_pred = lgbm_model.predict(X_test)

# 4. Evaluation
accuracy = accuracy_score(y_test, y_pred)
print(f"LightGBM Accuracy: {accuracy:.4f}")
print(f"LightGBM Classification Report:\n{classification_report(y_test, y_pred)}")
```
